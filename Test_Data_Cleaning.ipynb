```python
# get the csv at thisurl and load it to a df
google_sheet_url = "https://docs.google.com/spreadsheets/d/17m7t0zzuPEuVA-bN-qM4hcHP_D9vg8EDxBHR9JAP1jA/edit?usp=sharing"

# Extract the file ID from the Google Drive URL
file_id = google_sheet_url.split('/d/')[-1].split('/')[0]

# Construct the CSV export URL
csv_export_url = f"https://docs.google.com/spreadsheets/d/{file_id}/export?format=csv"

response = requests.get(csv_export_url)
response.raise_for_status()  # Raise an exception for HTTP errors

# Read CSV data into a pandas DataFrame with the first row as the header
df = pd.read_csv(io.StringIO(response.text), header=0)

print(f"Successfully loaded data from Google Sheet with {df.shape[0]} rows and {df.shape[1]} columns")
print(df.head())
```

```python
# prompt: # prompt: for each column, get the number of records where that column has i - zero values, ii- null values
# for column in df.columns:
#   print(f"Column: {column}")
#   print(f"  Number of zero values: {(df[column] == 0).sum()}")
#   print(f"  Number of null values: {df[column].isnull().sum()}")
# iplot these in a way i can compare them. put column names as x labels

import plotly.graph_objects as go

zero_counts = []
null_counts = []
column_names = []

for column in df.columns:
  column_names.append(column)
  zero_counts.append((df[column] == 0).sum())
  null_counts.append(df[column].isnull().sum())

fig = go.Figure(data=[
    go.Bar(name='Zero Values', x=column_names, y=zero_counts),
    go.Bar(name='Null Values', x=column_names, y=null_counts),
])

fig.update_layout(barmode='group',
                  title='Number of Zero, Null, and Empty Values per Column',
                  xaxis_title='Column Names',
                  yaxis_title='Count')

fig.show()
```

```python
# prompt: drop the unamed column

# Assuming 'df' is your DataFrame from the previous code

# Check if unnamed columns exist before attempting to drop them
if 'Unnamed: 0' in df.columns:
    df = df.drop('Unnamed: 0', axis=1)
    print("Dropped 'Unnamed: 0' column")

# Check if other unnamed columns exist
for col in df.columns:
    if 'Unnamed' in col:
        df = df.drop(col, axis=1)
        print(f"Dropped '{col}' column")

print(df.head())
```
